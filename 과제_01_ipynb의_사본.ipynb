## 칼라 영상으로부터 r channel, g channel, b channel 별로 구분한 후 각 채널별 256 단계로 히스트로그램 및 normalized histogram을 구하기 위한 함수를 histogram_256, normal_histogram_256의 이름을 가지도록 코드를 작성한 후 각 채널별 히스토그램을 가시화 하시오. 

```
r_channel = cx_train[40][:, :, 0]
g_channel = cx_train[40][:, :, 1]
b_channel = cx_train[40][:, :, 2]

def histogram_256(img):
     bins = np.zeros(256, np.int32)
     height, width = img.shape
     for i in range(0, height):
        for j in range(0, width):
             bins[img[i][j]] += 1

     return bins        
  
def normal_histogram_256(img):
    height, width = img.shape
    bins = histogram_256(img)
    bin = bins/(width*height)

    return bin

#히스토그램
bins = histogram_256(r_channel)
#정규화된 히스트그램
bin = normal_histogram_256(r_channel)

plt.figure(figsize=(16,8))
#Subplot 1
plt.subplot(1,2,1)
plt.plot(bins)
plt.xlim([0,256])
plt.title('R_Channel Histogram')
#Subplot 2
plt.subplot(1,2,2)
plt.plot(bin)
plt.xlim([0,256])
plt.title('R_Channel Normalized Histogram')
#Show plot
plt.show()


bins = histogram_256(g_channel)
bin = normal_histogram_256(g_channel)

plt.figure(figsize=(16,8))
#Subplot 1
plt.subplot(1,2,1)
plt.plot(bins)
plt.xlim([0,256])
plt.title('G_Channel Histogram')
#Subplot 2
plt.subplot(1,2,2)
plt.plot(bin)
plt.xlim([0,256])
plt.title('G_Channel Normalized Histogram')
#Show plot
plt.show()


bins = histogram_256(b_channel)
bin = normal_histogram_256(b_channel)

plt.figure(figsize=(16,8))
#Subplot 1
plt.subplot(1,2,1)
plt.plot(bins)
plt.xlim([0,256])
plt.title('B_Channel Histogram')
#Subplot 2
plt.subplot(1,2,2)
plt.plot(bin)
plt.xlim([0,256])
plt.title('B_Channel Normalized Histogram')
#Show plot
plt.show()
```

## 각 r channel, g channel, b channel별 16 단계로 단계로 히스트로그램 및 normalized histogram을 구하기 위한 함수를 histogram_256, normal_histogram_256의 이름을 가지도록 코드를 작성한 후 각 채널별 히스토그램을 가시화 하시오.

```
def histogram_16(img):
     bins = np.zeros(16, np.int32)
     height, width = img.shape
     for i in range(0, height):
        for j in range(0, width):
             bins[int(img[i][j]/16)] += 1

     return bins        
  
def normal_histogram_16(img):
    height, width = img.shape
    bins = histogram_16(img)
    bin = bins/(width*height)

    return bin
    
bins = histogram_16(r_channel)
bin = normal_histogram_16(r_channel)

plt.figure(figsize=(16,8))
#Subplot 1
plt.subplot(1,2,1)
plt.plot(bins)
plt.xlim([0,16])
plt.title('R_Channel Histogram')
#Subplot 2
plt.subplot(1,2,2)
plt.plot(bin)
plt.xlim([0,16])
plt.title('R_Channel Normalized Histogram')
#Show plot
plt.show()

bins = histogram_16(g_channel)
bin = normal_histogram_16(g_channel)

plt.figure(figsize=(16,8))
#Subplot 1
plt.subplot(1,2,1)
plt.plot(bins)
plt.xlim([0,16])
plt.title('G_Channel Histogram')
#Subplot 2
plt.subplot(1,2,2)
plt.plot(bin)
plt.xlim([0,16])
plt.title('G_Channel Normalized Histogram')
#Show plot
plt.show()

bins = histogram_16(b_channel)
bin = normal_histogram_16(b_channel)

plt.figure(figsize=(16,8))
#Subplot 1
plt.subplot(1,2,1)
plt.plot(bins)
plt.xlim([0,16])
plt.title('B_Channel Histogram')
#Subplot 2
plt.subplot(1,2,2)
plt.plot(bin)
plt.xlim([0,16])
plt.title('B_Channel Normalized Histogram')
#Show plot
plt.show()
```
